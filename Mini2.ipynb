Ejercicio 4: Repartidor de periódicos
    Decripción del problema y algoritmo
    

```python
def Periodicos(dias,p_comprados):
    p = [0.3,0.4,0.3]
    
    gananciaFinal = 0
    for i in range(dias):
        x = random.random()
        if (x < p[0]):
            p_solicitados = 9
        elif(x < p[0]+p[1]):
            p_solicitados = 10
        else:
            p_solicitados = 11

        p_vendidos = 0
        if(p_solicitados <= p_comprados):
            p_vendidos = p_solicitados
        else:
            p_vendidos = p_comprados

        p_sobrantes = p_comprados - p_vendidos

        inversion = p_comprados*1.5
        retorno = p_sobrantes*0.5
        ingreso = p_vendidos*2.5

        ganancia = ingreso - inversion + retorno
        gananciaFinal = gananciaFinal + ganancia
    
    return gananciaFinal
```

Descripción de los resultados finales

```python
# Simulaciones para 1 mes
print("Ganancia al comprar 9 periodicos por 1 mes",Periodicos(30,9))
print("Ganancia al comprar 10 periodicos por 1 mes",Periodicos(30,10))
print("Ganancia al comprar 11 periodicos por 1 mes",Periodicos(30,11))
# Simulaciones para 1 año
print("Ganancia al comprar 9 periodicos por 1 año",Periodicos(365,9))
print("Ganancia al comprar 10 periodicos por 1 año",Periodicos(365,10))
print("Ganancia al comprar 11 periodicos por 1 año",Periodicos(365,11))
# Simulaciones para 10 años
print("Ganancia al comprar 9 periodicos por 10 años",Periodicos(3652,9))
print("Ganancia al comprar 10 periodicos por 10 años",Periodicos(3652,10))
print("Ganancia al comprar 11 periodicos por 10 años",Periodicos(3652,11))
```
